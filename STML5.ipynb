```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
data = pd.read_csv('Salary_Data.csv')
features = data[['YearsExperience']]
target = data['Salary']
plt.scatter(features, target)
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
scaler = MinMaxScaler()
normalized_features = scaler.fit_transform(features)
normalized_target = scaler.fit_transform(target.values.reshape(-1, 1))
normalized_data = pd.DataFrame(normalized_features, columns=['Years of Experience'])
normalized_data['Salary'] = normalized_target
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
data = pd.read_csv('diabetes_data_upload.csv')
features = data.drop('class', axis=1)
target = data['class']
le = LabelEncoder()
for column in features.columns:
    if features[column].dtype == 'object':
        features[column] = le.fit_transform(features[column])
for feature in features.columns:
    plt.scatter(features[feature], target)
    plt.xlabel(feature)
    plt.ylabel('Class')
    plt.show()
scaler = MinMaxScaler()
normalized_features = scaler.fit_transform(features)
normalized_data = pd.DataFrame(normalized_features, columns=features.columns)
normalized_data['class'] = target
```
